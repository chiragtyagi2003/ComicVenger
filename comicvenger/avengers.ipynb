```python
# Join all strings of comics into one big string
comics_joined = "\n\n".join(comics)
```

```python
# find unique number of characters in the joined string
# this will help determine the length of our vocab
vocab = sorted(set(comics_joined))
```

```python
# vectorizing the text data

# create a mapping from char to unique index
# loop the list with enumerate and make key value pairs
# i => index
# u => char
char2idx = {u:i for i,u in enumerate(vocab)}
```

```python
# create a mapping from index to char
# just convert the vocab list to array
idx2char = np.array(vocab)
```

```python
# function to vectorize the comics string

def vectorize_string(string):
  vectorized_ouput = np.array([char2idx[char] for char in string])
  return vectorized_ouput
```

```python
# create training dataset
def get_batch(vectorized_comics, seq_length, batch_size):
  ''' function to generate training dataset'''

  # the length of vectorized string
  n = vectorized_comics.shape[0]-1

  # randomly choose  starting indices for training examples
  # number of indices to choose = batch_size

  # idx --> numpy array
  idx = np.random.choice(n-seq_length, batch_size)


  # construct a list of input sequences for traning batch
  input_batch = [vectorized_comics[i: i+seq_length] for i in idx]


  # construct a list of target sequences for training batch
  # this would be just 1 char to the right of input seq
  # to determine the next char
  output_batch = [vectorized_comics[i+1 : i+seq_length+1] for i in idx]

  # transform into x_batch and y_batch arrays
  x_batch = np.reshape(input_batch, [batch_size, seq_length])
  y_batch = np.reshape(output_batch, [batch_size, seq_length])


  return x_batch, y_batch
```

```python
# test the function
x_train_batch, y_train_batch = get_batch(vectorized_comics, 10, 2)

print(x_train_batch)
print('\n\n', y_train_batch)
```

```python
# function to add an LSTM layer

def LSTM(rnn_units):
  return tf.keras.layers.LSTM(
      rnn_units,
      return_sequences = True,
      recurrent_initializer = 'glorot_uniform',
      recurrent_activation = 'sigmoid',
      stateful = True,
  )
```

```python
# function to create a RNN model

def build_model(vocab_size, embedding_dim, rnn_units, batch_size):
  ''' function to add layers to the model and build it'''

  model = tf.keras.Sequential([

    # layer1 => embedding input layer
    # none in batch_input_shape allows the model to accept seqs of different lengths
    tf.keras.layers.Embedding(vocab_size, embedding_dim, batch_input_shape = [batch_size, None]),

    # layer2 => LSTM with rnn units
    LSTM(rnn_units),

    # layer3 => Dense layer -> this transforms the output of LSTM to vocab size
    tf.keras.layers.Dense(vocab_size)
  ])

  return model


```

```python
# defining the loss function
def compute_loss(labels, logits):
  loss = tf.keras.losses.sparse_categorical_crossentropy(labels, logits, from_logits = True)
  return loss
```

```python
# set hyperparameters

# optimization params
num_training_iterations = 1000
batch_size = 32
seq_length = 2000
learning_rate = 5e-3

# model params
vocab_size = len(vocab)
embedding_dim = 256
rnn_units = 1024


# checkpoint location
checkpoint_dir = './training_checkpoints'
checkpoint_prefix = os.path.join(checkpoint_dir, "my_ckpt")
```

```python
model = build_model(vocab_size, embedding_dim, rnn_units, batch_size)
```

```python
# training method
model = build_model(vocab_size, embedding_dim, rnn_units, batch_size)

optimizer = tf.keras.optimizers.Adam(learning_rate)

@tf.function
def train_step(x,y):
  with tf.GradientTape() as tape:


    # get predictions
    y_hat = model(x)

    # compute the loss
    loss = compute_loss(y, y_hat)

    # compute the gradient w.r.t all params
    grads = tape.gradient(loss, model.trainable_variables)

    # apply the grads to optimizer to optimize the weights
    optimizer.apply_gradients(zip(grads, model.trainable_variables))

    # return the loss
    return loss
```

```python
# begin training

# to store the opertaions while training
history = []
plotter = mdl.util.PeriodicPlotter(sec=2, xlabel = 'Iteration', ylabel = 'loss')

if hasattr(tqdm, '_instances'): tqdm._instances.clear() # clear if it exists

for iter in tqdm(range(num_training_iterations)):

  # grab a batch and propogate it forward through a network
  x_batch, y_batch = get_batch(vectorized_comics, seq_length, batch_size)

  loss = train_step(x_batch, y_batch)


  # update the progress bar
  history.append(loss.numpy().mean())
  plotter.plot(history)

  # update the model with the changed weights
  # after every 100 iterations
  if iter % 100 == 0:
    model.save_weights(checkpoint_prefix)


  # save the trained model and the weights
  model.save_weights(checkpoint_prefix)
```

```python
# rebuild the model to restore optimized weights from the last checkpoint

model = build_model(vocab_size, embedding_dim, rnn_units, batch_size = 1)
# restore the model weights for the last checkpoint after training
model.load_weights(tf.train.latest_checkpoint(checkpoint_dir))
model.build(tf.TensorShape([1, None]))


model.summary()
```

```python
# generate a new  comic stuff
def generate_text(model, start_string, generation_length = 1000):

  # vectorize the start string
  input_string = [char2idx[char] for char in start_string]

  # transforming string dims
  # as model takes 2 dim array
  input_string = tf.expand_dims(input_string, 0)

  # to store generated text
  text_generated = []

  model.reset_states()
  tqdm._instances.clear()

  for i in tqdm(range(generation_length)):

    # generate the next char
    predictions = model(input_string)

    # remove the batch dim
    predictions = tf.squeeze(predictions, 0)

    # sample the output using multinomial distrubution
    predicted_id = tf.random.categorical(predictions, num_samples = 1)[-1, 0].numpy()


    # pass the prediction along with the previous hidden state to model as the next
    # input
    input_string = tf.expand_dims([predicted_id], 0)

    # add the predicted char to the generated text
    text_generated.append(idx2char[predicted_id])


  return (start_string + ''.join(text_generated))


```
